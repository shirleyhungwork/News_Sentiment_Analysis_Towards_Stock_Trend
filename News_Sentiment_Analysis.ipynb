# Topic Modelling

```python
import pandas as pd
import os
df = pd.read_csv(os.getcwd()+'\\output\\news_database.csv')
```

```python
%%time
for i in ['tsla','amzn','googl','msft','adbe']:
  globals()[f'{i}_dt'] = stock(i)
  df[i] = eval(f'{i}_dt').detect(df)
```

```python
print(df[df['tsla']==1].title.tolist())
```

# News Sentiment

**Stock Price Trend Prediction**

```python
for s in ['tsla','amzn','googl','adbe','msft']:
  news = news_sentiment(df,s,'Price')
  for i in range(6):
    print(f'stock {s} day {i}')
    out = news.predict(i)
    print("========================== \n")
```
